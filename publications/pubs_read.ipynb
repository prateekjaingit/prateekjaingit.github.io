```python
import json
from serpapi import GoogleSearch

# Opening JSON file generated by serpapi playground, it should list all publications for a given author
f = open('pubs.json',)

# returns JSON object as
# a dictionary
pubs = json.load(f)

# Iterating through the json
# list
i=0
results=[]
for pub in pubs:
    #works only for 100 searches
    cit_id=pub['citation_id']
    params = {
        "engine": "google_scholar_author",
        "view_op": "view_citation",
        "citation_id": cit_id,
        "api_key": "b149bd5cbd6f123d25a5a6d0e3754a771307c999ee58a2dc33f7ab5593b18f9f"
    }
    search = GoogleSearch(params)
    results.append(search.get_dict())
    i=i+1


# Closing file
f.close()
```

```python
with open('pub_results.json', 'w') as f:
    json.dump(results, f)
```
